## Ejercicio 1
**Expresión a resolver:** `(7 + 2) * (5 - 3)`

### a. Construcción del Árbol

```
       *
      / \
     +   -
    / \ / \
   7  2 5  3
```


````python
# Definimos las conexiones entre nodos (estructura del árbol)
edges = [
    ('*', '+'),    # '*' tiene como hijo izquierdo '+'
    ('*', '-'),    # '*' tiene como hijo derecho '-'
    ('+', '7'),    # '+' tiene como hijo izquierdo '7'
    ('+', '2'),    # '+' tiene como hijo derecho '2'
    ('-', '5'),    # '-' tiene como hijo izquierdo '3'
    ('-', '3')     # '-' tiene como hijo derecho '4'
]


# Definimos las posiciones para una visualización jerárquica
pos = {
    '*': (0, 2),     # Raíz
    '+': (-1, 1),    # Hijo izquierdo de '*'
    '-': (1, 1),     # Hijo derecho de '*'
    '7': (-1.5, 0),  # Hijo izquierdo de '+'
    '2': (-0.5, 0),  # Hijo derecho de '+'
    '5': (0.5, 0),   # Hijo izquierdo de '-'
    '3': (1.5, 0)    # Hijo derecho de '-'
}
````

### b. Visualización con `networkx`

````python
# Escribe aquí el código para graficar el árbol
# Importamos las bibliotecas necesarias
import matplotlib.pyplot as plt
import networkx as nx

# Creamos un nuevo grafo dirigido (DiGraph)
G = nx.DiGraph()

# Definimos las conexiones entre nodos (estructura del árbol)
edges = [
    ('*', '+'),    # '*' tiene como hijo izquierdo '+'
    ('*', '-'),    # '*' tiene como hijo derecho '-'
    ('+', '7'),    # '+' tiene como hijo izquierdo '7'
    ('+', '2'),    # '+' tiene como hijo derecho '2'
    ('-', '5'),    # '-' tiene como hijo izquierdo '3'
    ('-', '3')     # '-' tiene como hijo derecho '4'
]

# Agregamos las aristas al grafo
G.add_edges_from(edges)

# Definimos las posiciones para una visualización jerárquica
pos = {
    '*': (0, 2),     # Raíz
    '+': (-1, 1),    # Hijo izquierdo de '*'
    '-': (1, 1),     # Hijo derecho de '*'
    '7': (-1.5, 0),  # Hijo izquierdo de '+'
    '2': (-0.5, 0),  # Hijo derecho de '+'
    '5': (0.5, 0),   # Hijo izquierdo de '-'
    '3': (1.5, 0)    # Hijo derecho de '-'
}

# Dibujamos el árbol
plt.figure(figsize=(10, 5))
nx.draw(
    G, pos,
    with_labels=True,
    node_color='lightblue',
    node_size=2000,
    font_size=14,
    font_weight='bold',
    arrows=True
)
plt.title("Árbol de expresión: (7 + 2) * (5 - 3)")
plt.axis('off')
plt.show()
````

### c. Evaluación de la Expresión

````python
# Escribe aquí el código para evaluar la expresión

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None
        self.der = None

def construir_arbol():
    raiz = Nodo('*')
    raiz.izq = Nodo('+')
    raiz.der = Nodo('-')
    raiz.izq.izq = Nodo(7)
    raiz.izq.der = Nodo(2)
    raiz.der.izq = Nodo(5)
    raiz.der.der = Nodo(3)
    return raiz

def evaluar(nodo):
    if isinstance(nodo.valor, int):
        return nodo.valor
    izq = evaluar(nodo.izq)
    der = evaluar(nodo.der)
    if nodo.valor == '+':
        return izq + der
    elif nodo.valor == '-':
        return izq - der
    elif nodo.valor == '*':
        return izq * der
    elif nodo.valor == '/':
        return izq / der

# Construimos y evaluamos el árbol
arbol = construir_arbol()
resultado = evaluar(arbol)
resultado





````

````python
# Escribe aquí el código para evaluar la expresión
def construir_arbol():
    raiz = Nodo('/')
    raiz.izq = Nodo('+')
    raiz.der = Nodo('-')

    # Subárbol izquierdo: (5 + (2 * (8 - 3)))
    raiz.izq.izq = Nodo(5)
    raiz.izq.der = Nodo('*')
    raiz.izq.der.izq = Nodo(2)
    raiz.izq.der.der = Nodo('-')
    raiz.izq.der.der.izq = Nodo(8)
    raiz.izq.der.der.der = Nodo(3)

    # Subárbol derecho: (6 - (1 + 2))
    raiz.der.izq = Nodo(6)
    raiz.der.der = Nodo('+')
    raiz.der.der.izq = Nodo(1)
    raiz.der.der.der = Nodo(2)

    return raiz

def evaluar(nodo):
    if isinstance(nodo.valor, int):
        return nodo.valor
    izq = evaluar(nodo.izq)
    der = evaluar(nodo.der)
    if nodo.valor == '+':
        return izq + der
    elif nodo.valor == '-':
        return izq - der
    elif nodo.valor == '*':
        return izq * der
    elif nodo.valor == '/':
        return izq / der

# Construimos y evaluamos el árbol
arbol = construir_arbol()
resultado = evaluar(arbol)
resultado


````
