## üß™ Tarea 1: √Årbol binario de 7 nodos

Valores insertados: **50, 30, 70, 20, 40, 60, 80**

Representaci√≥n del √°rbol binario:

```
        50         ‚Üê ra√≠z (nivel 0)
       /  \
     30    70      ‚Üê nivel 1
    / \    / \
  20  40  60  80   ‚Üê nivel 2
```

### Identificaci√≥n de elementos:

| Elemento       | Valor(es)                              |
|----------------|----------------------------------------|
| **Ra√≠z**       | 50                                     |
| **Hojas**      | 20, 40, 60, 80                         |
| **Nodos internos** | 30, 70                              |
| **Nivel 0**    | 50                                     |
| **Nivel 1**    | 30, 70                                 |
| **Nivel 2**    | 20, 40, 60, 80                         |
| **Altura**     | 2                                      |


````python
# Importamos la librer√≠a matplotlib para generar gr√°ficos (plt es el alias com√∫n).
import matplotlib.pyplot as plt

# Importamos la librer√≠a networkx para trabajar con grafos dirigidos (nx es el alias com√∫n).
import networkx as nx

# Creamos un nuevo grafo dirigido (DiGraph significa "Directed Graph", es decir, con flechas).
G = nx.DiGraph()

# Definimos las conexiones (aristas) entre los nodos. Cada tupla representa una relaci√≥n padre ‚Üí hijo.
edges = [
    (50, 30),  # El nodo ra√≠z 50 tiene como hijo izquierdo al 30
    (50, 70),  # El nodo ra√≠z 50 tiene como hijo derecho al 70
    (30, 20),  # El nodo 30 tiene como hijo izquierdo al 20
    (30, 40),  # El nodo 30 tiene como hijo derecho al 40
    (70, 60),  # El nodo 70 tiene como hijo izquierdo al 60
    (70, 80)   # El nodo 70 tiene como hijo derecho al 80
]

# Agregamos todas las aristas al grafo G. NetworkX crear√° autom√°ticamente los nodos implicados.
G.add_edges_from(edges)

# Definimos las posiciones en el plano (x, y) para que el √°rbol se visualice de forma jer√°rquica.
pos = {
    50: (0, 2),         # Ra√≠z del √°rbol, nivel 0
    30: (-1.5, 1),      # Hijo izquierdo de 50, nivel 1
    70: (1.5, 1),       # Hijo derecho de 50, nivel 1
    20: (-2.2, 0),      # Hijo izquierdo de 30, nivel 2
    40: (-0.8, 0),      # Hijo derecho de 30, nivel 2
    60: (0.8, 0),       # Hijo izquierdo de 70, nivel 2
    80: (2.2, 0)        # Hijo derecho de 70, nivel 2
}

# Creamos una nueva figura con tama√±o de 10x6 pulgadas
plt.figure(figsize=(10, 6))

# Dibujamos el grafo con las siguientes caracter√≠sticas:
# - G: el grafo que creamos
# - pos: las coordenadas (x, y) de cada nodo
# - with_labels=True: para mostrar los valores de cada nodo
# - node_color='lightgreen': color de los nodos
# - node_size=1800: tama√±o de cada nodo
# - font_size=14: tama√±o del texto
# - font_weight='bold': texto en negrita
# - arrows=False: no mostrar flechas (por ser √°rbol binario sin orientaci√≥n visual)
nx.draw(
    G, pos,
    with_labels=True,
    node_color='lightgreen',
    node_size=1800,
    font_size=14,
    font_weight='bold',
    arrows=False
)

# T√≠tulo del gr√°fico
plt.title("√Årbol Binario de B√∫squeda: [50, 30, 70, 20, 40, 60, 80]", fontsize=14)

# Oculta los ejes del gr√°fico
plt.axis('off')

# Muestra el gr√°fico en pantalla
plt.show()
````

## üß™ Tarea 2 (opcional): √Årbol binario en expresiones aritm√©ticas

### Expresi√≥n: `(3 + 5) * (2 - 1)`

√Årbol de expresi√≥n:

```
       *
      / \
     +   -
    / \ / \
   3  5 2  1
```

### Aplicaciones:
- Evaluaci√≥n autom√°tica de expresiones
- An√°lisis sint√°ctico en compiladores
- Conversi√≥n entre notaciones (infix, postfix, prefix)

Cada hoja es un operando (n√∫mero), y cada nodo interno es un operador.


````python
# Importamos la librer√≠a matplotlib para generar gr√°ficos (plt es el alias com√∫n).
import matplotlib.pyplot as plt

# Importamos la librer√≠a networkx para trabajar con grafos dirigidos (nx es el alias com√∫n).
import networkx as nx

# Creamos un nuevo grafo dirigido (DiGraph significa "Directed Graph", es decir, con flechas).
G = nx.DiGraph()

# Definimos las conexiones (aristas) entre los nodos. Cada tupla representa una conexi√≥n de padre a hijo.
edges = [
    ('*', '+'),  # El nodo ra√≠z '*' tiene como hijo izquierdo el operador '+'
    ('*', '-'),  # El nodo ra√≠z '*' tiene como hijo derecho el operador '-'
    ('+', '3'),  # El nodo '+' tiene como hijo izquierdo el operando '3'
    ('+', '5'),  # El nodo '+' tiene como hijo derecho el operando '5'
    ('-', '2'),  # El nodo '-' tiene como hijo izquierdo el operando '2'
    ('-', '1')   # El nodo '-' tiene como hijo derecho el operando '1'
]

# Agregamos todas las aristas al grafo G. NetworkX crear√° autom√°ticamente los nodos implicados.
G.add_edges_from(edges)

# Definimos las posiciones en el plano (x, y) para que el √°rbol se visualice ordenado y jer√°rquico.
pos = {
    '*': (0, 2),      # La ra√≠z est√° en el nivel m√°s alto
    '+': (-1, 1),     # Hijo izquierdo de '*'
    '-': (1, 1),      # Hijo derecho de '*'
    '3': (-1.5, 0),   # Hijo izquierdo de '+'
    '5': (-0.5, 0),   # Hijo derecho de '+'
    '2': (0.5, 0),    # Hijo izquierdo de '-'
    '1': (1.5, 0)     # Hijo derecho de '-'
}

# Creamos una nueva figura de tama√±o 8x5 pulgadas
plt.figure(figsize=(8, 5))

# Dibujamos el grafo:
# - G: el grafo a dibujar
# - pos: las posiciones definidas para cada nodo
# - with_labels=True: muestra los nombres (etiquetas) de los nodos
# - node_color='lightblue': color de los nodos
# - node_size=2000: tama√±o de los nodos
# - font_size=14, font_weight='bold': tama√±o y grosor del texto
# - arrows=True: muestra flechas indicando direcci√≥n
nx.draw(
    G, pos,
    with_labels=True,
    node_color='lightblue',
    node_size=2000,
    font_size=14,
    font_weight='bold',
    arrows=True
)

# T√≠tulo del gr√°fico
plt.title("√Årbol de expresi√≥n: (3 + 5) * (2 - 1)")

# Muestra el gr√°fico en pantalla
plt.show()
````
