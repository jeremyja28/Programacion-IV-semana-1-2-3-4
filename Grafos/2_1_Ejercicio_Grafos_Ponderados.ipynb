## Ejercicio: Grafo de rutas entre ciudades

### Enunciado

Una empresa de transporte desea modelar un sistema de rutas entre ciudades para estimar los costos de viaje. El grafo se compone de las siguientes ciudades:

- Quito
- Ambato
- Riobamba
- Cuenca

Y las rutas con sus respectivos costos (en USD):

- Quito ‚Üí Ambato: 10
- Quito ‚Üí Riobamba: 18
- Ambato ‚Üí Cuenca: 25
- Riobamba ‚Üí Cuenca: 20

### Objetivo del estudiante

1. Crear un grafo dirigido y ponderado con `networkx`.
2. Representar el grafo visualmente.
3. Mostrar los pesos de las rutas.
4. Determinar el **camino m√°s barato** entre Quito y Cuenca usando un algoritmo como `shortest_path` o `dijkstra_path`.

```python
#1. Crear un grafo dirigido y ponderado con `networkx`.
# üì¶ Importamos librer√≠as necesarias
import networkx as nx
import matplotlib.pyplot as plt

# üß† Creamos un grafo dirigido y ponderado
G = nx.DiGraph()

# üß± A√±adimos nodos manualmente (opcional, tambi√©n se crean autom√°ticamente al agregar aristas)
G.add_nodes_from(['Quito', 'Ambato', 'Riobamba', 'Cuenca'])

# ‚ûï A√±adimos aristas con pesos usando el par√°metro 'weight'
G.add_edge('Quito', 'Ambato', weight=10)
G.add_edge('Quito', 'Riobamba', weight=18)
G.add_edge('Ambato', 'Cuenca', weight=25)
G.add_edge('Riombamba', 'Cuenca', weight=20)
print("Grado de cada nodo:")
# Grado de cada nodo
for nodo in G.nodes:
    print(f"{nodo}: entrada = {G.in_degree(nodo)}, salida = {G.out_degree(nodo)}")
```

```python
#2. Representar el grafo visualmente.
# üì¶ Importamos librer√≠as necesarias
import networkx as nx
import matplotlib.pyplot as plt

# üß† Creamos un grafo dirigido y ponderado
G = nx.DiGraph()

# üß± A√±adimos nodos manualmente (opcional, tambi√©n se crean autom√°ticamente al agregar aristas)
G.add_nodes_from(['Quito', 'Ambato', 'Riobamba', 'Cuenca'])

# ‚ûï A√±adimos aristas con pesos usando el par√°metro 'weight'
G.add_edge('Quito', 'Ambato', weight=10)
G.add_edge('Quito', 'Riobamba', weight=18)
G.add_edge('Ambato', 'Cuenca', weight=25)
G.add_edge('Riobamba', 'Cuenca', weight=20)

# üéØ Posiciones para graficar el grafo (usa spring_layout para distribuci√≥n est√©tica)
pos = nx.spring_layout(G, seed=62)

# üé® Dibujamos nodos y aristas
plt.figure(figsize=(8, 5))
nx.draw(
    G, pos, with_labels=True,
    node_color='lightblue', node_size=2000,
    font_size=14, font_weight='bold', arrows=True
)

# üè∑Ô∏è Extraemos los pesos de las aristas para dibujarlos como etiquetas
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# üßæ T√≠tulo
plt.title("Grafo de Rutas entre Ciudades")
plt.axis('off')
plt.show()
```

```python
#3. Mostrar los pesos de las rutas.
# üì¶ Importamos librer√≠as necesarias
import networkx as nx
import matplotlib.pyplot as plt

# üß† Creamos un grafo dirigido y ponderado
G = nx.DiGraph()

# üß± A√±adimos nodos manualmente (opcional, tambi√©n se crean autom√°ticamente al agregar aristas)
G.add_nodes_from(['Quito', 'Ambato', 'Riobamba', 'Cuenca'])

# ‚ûï A√±adimos aristas con pesos usando el par√°metro 'weight'
G.add_edge('Quito', 'Ambato', weight=10)
G.add_edge('Quito', 'Riobamba', weight=18)
G.add_edge('Ambato', 'Cuenca', weight=25)
G.add_edge('Riobamba', 'Cuenca', weight=20)

# üéØ Posiciones para graficar el grafo (usa spring_layout para distribuci√≥n est√©tica)
pos = nx.spring_layout(G, seed=62)

# üé® Dibujamos nodos y aristas
plt.figure(figsize=(8, 5))
nx.draw(
    G, pos, with_labels=True,
    node_color='lightblue', node_size=2000,
    font_size=14, font_weight='bold', arrows=True
)

# üè∑Ô∏è Extraemos los pesos de las aristas para dibujarlos como etiquetas
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# üßæ T√≠tulo
plt.title("Grafo de Rutas entre Ciudades")
plt.axis('off')
plt.show()

# Mostrar todos los pesos asignados
for u, v, datos in G.edges(data=True):
    print(f"{u} ‚Üí {v} | peso = {datos['weight']}")
```

```python
#4. Determinar el **camino m√°s barato** entre Quito y Cuenca usando un algoritmo como `shortest_path` o `dijkstra_path`.
# üì¶ Importamos librer√≠as necesarias
import networkx as nx
import matplotlib.pyplot as plt

# üß† Creamos un grafo dirigido y ponderado
G = nx.DiGraph()

# üß± A√±adimos nodos manualmente (opcional, tambi√©n se crean autom√°ticamente al agregar aristas)
G.add_nodes_from(['Quito', 'Ambato', 'Riobamba', 'Cuenca'])

# ‚ûï A√±adimos aristas con pesos usando el par√°metro 'weight'
G.add_edge('Quito', 'Ambato', weight=10)
G.add_edge('Quito', 'Riobamba', weight=18)
G.add_edge('Ambato', 'Cuenca', weight=25)
G.add_edge('Riobamba', 'Cuenca', weight=20)

# üéØ Posiciones para graficar el grafo (usa spring_layout para distribuci√≥n est√©tica)
pos = nx.spring_layout(G, seed=62)

# üé® Dibujamos nodos y aristas
plt.figure(figsize=(8, 5))
nx.draw(
    G, pos, with_labels=True,
    node_color='lightblue', node_size=2000,
    font_size=14, font_weight='bold', arrows=True
)

# üè∑Ô∏è Extraemos los pesos de las aristas para dibujarlos como etiquetas
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# üßæ T√≠tulo
plt.title("Grafo de Rutas entre Ciudades")
plt.axis('off')
plt.show()
# üß† Encontramos el camino m√°s barato y su costo total usando Dijkstra
camino = nx.dijkstra_path(G, source='Quito', target='Cuenca')
costo = nx.dijkstra_path_length(G, source='Quito', target='Cuenca')

# Mostramos resultados
print("Camino m√°s barato de Quito a Cuenca:", camino)
print("Costo total del viaje:", costo)
```

El algoritmo de Dijkstra se emplea para encontrar la ruta m√°s econ√≥mica desde Quito hasta Cuenca, analizando los caminos posibles y seleccionando el de menor costo total.
